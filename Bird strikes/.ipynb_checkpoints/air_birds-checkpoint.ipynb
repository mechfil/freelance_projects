## Описание проекта

**Описание проекта**:  
Необходимо изучить влияние погоды на столкновения птиц с воздушными судами. Проанализировать данные для проведения дальнейшего исследования и выявить закономерности в данных. На основе исследования будет построена модель предсказания столкновений птиц с самолётами коллегами из DS-направления.

**Задачи проекта:** 
- Провести предобработку и отчистку данных
- Провести EDA-анализ данных
- Выявить параметры погоды, которые влияют на столкновения воздушных судов с птицами

**Рабочие файлы (приложены к проекту)**:  
- [data_DE_strikes_and_weather.csv](data_DE_strikes_and_weather.csv) - объединенный датасет столкновений и погоды
- [flights_data_DE.csv](flights_data_DE.csv) - датасет с рейсами для топ-22 аэропортов по числу столкновений

**Документация (приложена к проекту)**:
- [fieldlist.pdf](fieldlist.pdf) - документация с описаниями столбцов столкновений датасета **data_DE_strikes_and_weather.csv**
- [st_data_info_NCEI.pdf](st_data_info_NCEI.pdf) - документация с описаниями погодных столбцов датасета **data_DE_strikes_and_weather.csv**
- [isd-format-document.pdf](isd-format-document.pdf) - дополнительная документация с описаниями погодных столбцов датасета **data_DE_strikes_and_weather.csv**
- [icao_report.pdf](icao_report.pdf) - документация с отчетом ICAO о столкновениях с животными воздушных судов и кодами видов животных
- [icao_manual.pdf](icao_manual.pdf) - документация ICAO по заполнению отчета о столкновении воздушных судов с животным

**Методы исследования**:  
- Предобработка данных
- EDA-анализ данных

## Описание данных

Датасет **data_DE_strikes_and_weather.csv**.

- Данный датасет получен объединением данных о столкновениях с животными воздушных судов от FAA (Federal Aviation Administration) https://www.faa.gov/ и данных о погоде от NCEI (National Centers for Environmental Information) https://www.ncei.noaa.gov/. Данные о погоде брались с ближайшей к аэропорту станции, разница составляет не более 1 часа. Данные о погоде не подгружались, если не указаны координаты и/или время столкновения. Записи страйков с 01/01/2018 по 30/08/2023. Если не указано время, то данные за 12:00

- Подробное описание столбцов датасета **data_DE_strikes_and_weather.csv** находится в приложенных к проекту файлах [fieldlist.pdf](fieldlist.pdf), [st_data_info_NCEI.pdf](st_data_info_NCEI.pdf) и [isd-format-document.pdf](isd-format-document.pdf)
- Краткое описание столбцов датасета **data_DE_strikes_and_weather.csv**, переведенное на русский язык:
- Столбцы данных о столкновениях животных с воздушными судами от FAA:
    - `Column1` - Столбец индекса строки
    - `INDEX_NR` - Индивидуальный номер записи
    - `INCIDENT_DATE` - Дата инцидента

    - `AMO` - Код модели воздушного судна по международной организации гражданской авиации 
    - `EMA` - Код производителя двигателя  
    - `EMO` - Код модели двигателя 
    - `AC_CLASS` - Тип воздушного судна 
    - `AC_MASS` - Масса воздушного судна 
    - `TYPE_ENG` - Тип двигателя

Зададим собственную функцию для обзора данных

```python
# функция для обзора столбцов датафрейма

def data_info(data_import):
    
    # создаем столбцы с нужными данными и склеиваем их в один датафрейм
    data_info = (pd.DataFrame(data_import.dtypes,
                              columns=['Тип столбца'])
             .join(pd.DataFrame(data_import.nunique(),
                                columns=['Уникальных значений']))
             .join(pd.DataFrame(data_import.isna().sum(),
                                columns=['Число пропусков']))
             .join(pd.DataFrame(data_import.isna().sum()/data_import.shape[0],
                                columns=['Процент пропусков'])))
    
    # функция для форматирования стилей с прозрачностью
    def highlight_cells(value):
        if value == 0:
            return 'background-color: rgba(0, 255, 0, 0.3)' 
        elif value > 0 and value <= 0.05:
            return 'background-color: rgba(255, 255, 0, 0.3)'
        elif value > 0.05:
            return 'background-color: rgba(255, 0, 0, 0.3)'
        else:
            return ''

    # применение стилей с прозрачностью к колонке с процентом пропусков
    display(data_info.style.applymap(lambda x: highlight_cells(x), 
                                     subset=pd.IndexSlice[:, ['Процент пропусков']])
            .format({'Процент пропусков': '{:.2%}'}))
    

# функция для обзора датафрейма и вывода первых строк

def data_review (data_import):
    
    # выводем случайные 5 строк
    print(display(data_import))
    
    # выведем общую обзорную информацию о датафрейме  
    display(pd.DataFrame(data=[data_import.shape[0],
                               data_import.shape[1],
                               data_import.duplicated().sum()],
                         columns=['Всего в датасете'],
                         index=['Строк','Столбцов','Явных дубликатов'])) 
    
    # выводем информацию о столбцах датафрейма
    data_info(data_import)
```

#### Добавление столбца с классом животного

Добавим столбец с обозначением класса животного. Соотношение ID животных и их классов возьмем из документации [icao_report.pdf](icao_report.pdf)

```python
# задаем функцию для определения класса животного
def func_spec_class(x):
    if x.startswith('1'): return 'MAMMAL'
    elif  x.startswith('2'): return 'REPTILE'
    elif  re.match('^[A-Za-z]', x): return 'BIRD'
    else: return '<unknown>'
    
# добавляем столбец с классом животного
df_sw['ANIMAL_CLASS'] = df_sw['SPECIES_ID'].astype('str').apply(func_spec_class)

# посмотрим результат
df_sw['ANIMAL_CLASS'].value_counts()
```

#### Добавление столбца с видами птиц

Добавим отдельный столбец с видами птиц. Соотношение видов птиц и их ID возьмем из документации [icao_report.pdf](icao_report.pdf)

```python
# задаем функцию для определния типа животного
def func_spec_type(x):
    if x.startswith('1'): return 'NOT BIRD'
    elif  x.startswith('2'): return 'NOT BIRD'
    elif  x.startswith('3'): return 'NOT BIRD'
    elif  x.startswith('C'): return 'KIWI'
    elif  x.startswith('F'): return 'GREBES'
    elif  x.startswith('G'): return 'ALBATROSS, SHEARWATERS, PETREL'
    elif  x.startswith('H'): return 'PELICANS, CORMORANTS, BOOBIES'
    elif  x.startswith('I'): return 'HERON, STORK, IBIS, FLAMINGO'
    elif  x.startswith('J'): return 'DUCKS, GEESE, SWANS'
    elif  x.startswith('K'): return 'HAWKS, EAGLES, VULTURES'
    elif  x.startswith('L'): return 'CHICKEN-LIKE BIRDS'
    elif  x.startswith('M'): return 'CRANES, RAILS'
    elif  x.startswith('N'): return 'SHORE BIRDS'
    elif  x.startswith('O'): return 'PIGEONS, GROUSE'
    elif  x.startswith('P'): return 'PARROT, MACAW, PARAKEET, LORIE'
    elif  x.startswith('Q'): return 'CUCKOOS'
    elif  x.startswith('R'): return 'OWLS'
    elif  x.startswith('S'): return 'GOATSUCKER NIGHTJAR FROGMOUTH'
    elif  x.startswith('T'): return 'SWIFT, TREE-SWIFT, HUMMINGBIRD'
    elif  x.startswith('W'): return 'KINGFISHERS, MOTMOTS, HORNBILL'
    elif  x.startswith('X'): return 'WOODPECKERS, TOUCANS, BARBETS'
    elif  x.startswith('Y') or x.startswith('Z'): return 'PERCHING BIRDS'
    elif  x.startswith('UNKB'): return 'UNKNOWN BIRDS'
    else: return '<unknown>'

# создаем столбец с видом птицы
df_sw['BIRD_GROUP'] = df_sw['SPECIES_ID'].astype('str').apply(func_spec_type)
```

### Формирование датасета с агрегированными данными по столкновениям, рейсам и параметрам погоды

Датасет со страйками с птицами далее нам необходимо объединить с датасетом с рейсами, поэтому проведем агрегацию, аналогичную агрегации в датасете с рейсами и добавим средние значения параметров погоды.

Сгруппируем данные по столбцам:
- Год
- Месяц
- Аэропорт

Добавим столбец с количеством страйков `COUNT_STRICES`, и добавим средние значения параметров погоды.

```python
# отбираем только птиц
df_sw_burds = df_sw[df_sw['ANIMAL_CLASS'] == 'BIRD']

# добавим столбец со страйком
df_sw_burds['STRIKES'] = 1

# агрегируем по столбцам
df_sw_burds_agg = (df_sw_burds.groupby(['INCIDENT_YEAR','INCIDENT_MONTH','AIRPORT_ID'])
                .agg({
                    'STRIKES':'sum',
                    'WND_ANG':'mean',
                    'WND_SPD':'mean',
                    'CIG':'mean',
                    'VIS':'mean',
                    'TMP':'mean',
                    'DEW':'mean',
                    'SLP':'mean',
                })
                .reset_index().rename(columns={
                    'WND_ANG':'WND_ANG_MEAN',
                    'WND_SPD':'WND_SPD_MEAN',
                    'CIG':'CIG_MEAN',
                    'VIS':'VIS_MEAN',
                    'TMP':'TMP_MEAN',
                    'DEW':'DEW_MEAN',
                    'SLP':'SLP_MEAN',
                })
               .round(2).sort_values(by=['INCIDENT_YEAR','INCIDENT_MONTH','AIRPORT_ID']))
```

Необходимо объединить датасет с рейсами и датасет со страйками. Перед объединением необходимо еще немного предобработать данные.

ID аэропортов в датасетах не совпадают. Проблема в несколько различающейся системе присвоения ID. В датасете со страйками ID аэропортов идут с буквой 'K' вначале. Добавим первую букву 'K' в ID аэропортов датасета с рейсами.

```python
# меняем ID аэропортов
df_f['airport'] = df_f['airport'].apply(lambda x: 'K' + x)
```

Чтобы присединить таблицы, нужно совпадение типов данных в столбцах присоединения. Поменяем тип столбца с месяцем в датасете с рейсами и страйками.

```python
# меняем тип столбца с месяцем в датасете с рейсами
df_f['Month'] = df_f['Month'].astype('int')

# меняем тип столбца с месяцем в датасете со страйками
df_sw_burds_agg['INCIDENT_MONTH'] = df_sw_burds_agg['INCIDENT_MONTH'].astype('int')
```

Теперь присоединим к датасету с рейсами датасет со страйками. Если после присоединения образуются пропуски в страйках - это значит, что в этом году и месяце в данном аэропорту не было страйков, поэтому заполним такие пропуски нулями. И в названиях аэропортов уберем букву K.

```python
# объединяем датасеты со страйками и рейсами
df_final = (df_f.merge(df_sw_burds_agg,
                      left_on=['Year','Month','airport'],
                      right_on=['INCIDENT_YEAR','INCIDENT_MONTH','AIRPORT_ID'], how='left')
            .drop(['AIRPORT_ID','INCIDENT_YEAR','INCIDENT_MONTH'], axis=1)
           .rename(columns={'TOTAL':'FLIGHTS',
                            'Year':'YEAR',
                            'Month':'MONTH',
                            'airport':'AIRPORT'
                           }).
            sort_values(by=['YEAR','MONTH','AIRPORT']).reset_index(drop=True)
           )

# заполняем пропуски в страйках нулями
df_final['STRIKES'] = df_final['STRIKES'].fillna(0).astype('int64')

# убираем букву K в названиях аэропортов
df_final['AIRPORT'] = df_final['AIRPORT'].str[1:]
```

#### Распределение столкновений по частям воздушного судна

Сформируем датасет с пораженными частами воздущного судна

```python
# составим список столбцов, которые нужно оставить
columns_str = [
    'STR_RAD',
    'STR_WINDSHLD',
    'STR_NOSE',
    'STR_ENG1',
    'STR_ENG2',
    'STR_ENG3',
    'STR_ENG4',
    'STR_PROP',
    'STR_WING_ROT',
    'STR_FUSE',
    'STR_LG',
    'STR_TAIL',
    'STR_LGHTS',
    'STR_OTHER',
    'ANIMAL_CLASS'
]

# оставляем нужные столбцы
df_str = df_sw[columns_str]

# переименуем столбцы
df_str.rename(columns={
    'STR_RAD': 'RADOME',
    'STR_WINDSHLD': 'WINDSHIELD',
    'STR_NOSE': 'NOSE',
    'STR_ENG1': 'ENGINE 1',
    'STR_ENG2': 'ENGINE 2',
    'STR_ENG3': 'ENGINE 3',
    'STR_ENG4': 'ENGINE 4',
    'STR_PROP': 'PROPELLER',
    'STR_WING_ROT': 'WING AND ROTOR',
    'STR_FUSE': 'FUSELAGE',
    'STR_LG': 'LANDING GEAR',
    'STR_TAIL': 'TAIL',
    'STR_LGHTS': 'LIGHTS',
    'STR_OTHER': 'OTHER',
}, inplace=True)
```

### Проверка гипотез о связях параметров погоды и столкновений

Проверим гипотезы о связях параметров погоды и столкновений

**Гипотеза 1:** Направление ветра влияет на число столкновений: направление ветра может менять направление миграций птиц

**Гипотеза 2:** Скорость ветра влияет на число столкновений: более высокие скорости ветра могут затруднять полеты птиц, что может приводить к уменьшению числа столкновений

**Гипотеза 3:** Высота нижнего слоя облаков влияет на число столкновений: более низкая граница нижнего слоя облаков может приводить к увеличению числа столкновений.

**Гипотеза 4:** Видимость влияет на число столкновений: более низкая видимость может приводить к увеличению числа столкновений.

**Гипотеза 5:** Температура воздуха влияет на число столкновений: более теплые месяцы могут приводить к увеличению числа столкновений.

**Гипотеза 6:** Точка росы влияет на число столкновений: месяцы с более высокой точкой росы (а следовательно более высокой температурой) могут приводить к увеличению числа столкновений.

**Гипотеза 7:** Давление воздуха влияет на число столкновений: давление воздуха может менять направление миграций птиц

Вначале посмотрим на корреляции между числом столкновений и отобранными параметрами погоды по всем данным

```python
# строим хитмеп
ax = sns.heatmap(df_final_update.corr(), vmin=-1, vmax=1,
            cmap= 'coolwarm',annot=True, fmt='.1f', linecolor='black',
        linewidths=0.01)
plt.title('Матрица коэффициентов корреляций Пирсона\n')
plt.show()
```

**Выводы:**
- Уверенных корреляций, кроме корреляции температуры и точки росы нет. Далее попробуем проверить гипотезы отдельно по аэропортам

#### Проверка гипотез о связи столкновений с птицами и параметров погоды по аэропортам

Чтобы проверить гипотезы, построим корреляционную матрицу по 21 аэропорту

```python
# создаем пустой датафрейм
df = pd.DataFrame(columns=df_final_update.columns[5:])
df = df.rename_axis('AIRPORT', axis=0)
p = 0

# формируем таблицу корреляций
for airport in list(df_final_update['AIRPORT'].unique()):
    for name_column in df.columns:
        df_155 = df_final_update[df_final_update['AIRPORT'] == airport]
        df.loc[p, name_column] = df_155['PERC_STRIKES'].corr(df_155[name_column])
    df = df.rename(index={p: airport})
    p += 1

# переводим данные в числовой тип
df = df.apply(pd.to_numeric)

# задаем подложку для графика
plt.figure(figsize=(10, 6))
ax = sns.heatmap(df, vmin=-1, vmax=1,
            cmap= 'coolwarm',annot=True, fmt='.1f', linecolor='black',
        linewidths=0.01)

# настраиваем параметры графика
plt.title('Коэффициенты корреляций Пирсона между процентом столкновений\n и параметрами погоды по аэропортам\n')
ax.xaxis.set_ticks_position('top')
ax.set_facecolor('#464544')

# выводим график
plt.show()
```

**Вывод:**

- Стокновения с рептилиями приводят либо к другим влияниям на полет (Other)  — \~80%, либо к прерванному взлету (Aborted Take-off)  — \~20%
- Столкновения рептилий чаще всего происходят с шасси самолета (Landing Gear) — (\~80%)
- Столкновения с рептилиями по имеющимся данным не приводили к повреждениям самолетов
