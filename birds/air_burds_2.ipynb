```python
# —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –æ–±–∑–æ—Ä–∞ —Å—Ç–æ–ª–±—Ü–æ–≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞

def data_info(data_import):
    
    # —Å–æ–∑–¥–∞–µ–º —Å—Ç–æ–ª–±—Ü—ã —Å –Ω—É–∂–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏ –∏ —Å–∫–ª–µ–∏–≤–∞–µ–º –∏—Ö –≤ –æ–¥–∏–Ω –¥–∞—Ç–∞—Ñ—Ä–µ–π–º
    data_info = (pd.DataFrame(data_import.dtypes,
                              columns=['–¢–∏–ø —Å—Ç–æ–ª–±—Ü–∞'])
             .join(pd.DataFrame(data_import.nunique(),
                                columns=['–£–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π']))
             .join(pd.DataFrame(data_import.isna().sum(),
                                columns=['–ß–∏—Å–ª–æ –ø—Ä–æ–ø—É—Å–∫–æ–≤']))
             .join(pd.DataFrame(data_import.isna().sum()/data_import.shape[0],
                                columns=['–ü—Ä–æ—Ü–µ–Ω—Ç –ø—Ä–æ–ø—É—Å–∫–æ–≤'])))
    
    # —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è —Ñ–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–∏—è —Å—Ç–∏–ª–µ–π —Å –ø—Ä–æ–∑—Ä–∞—á–Ω–æ—Å—Ç—å—é
    def highlight_cells(value):
        if value == 0:
            return 'background-color: rgba(0, 255, 0, 0.3)' 
        elif value > 0 and value <= 0.05:
            return 'background-color: rgba(255, 255, 0, 0.3)'
        elif value > 0.05:
            return 'background-color: rgba(255, 0, 0, 0.3)'
        else:
            return ''

    # –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ —Å—Ç–∏–ª–µ–π —Å –ø—Ä–æ–∑—Ä–∞—á–Ω–æ—Å—Ç—å—é –∫ –∫–æ–ª–æ–Ω–∫–µ —Å –ø—Ä–æ—Ü–µ–Ω—Ç–æ–º –ø—Ä–æ–ø—É—Å–∫–æ–≤
    display(data_info.style.applymap(lambda x: highlight_cells(x), 
                                     subset=pd.IndexSlice[:, ['–ü—Ä–æ—Ü–µ–Ω—Ç –ø—Ä–æ–ø—É—Å–∫–æ–≤']]).format({'–ü—Ä–æ—Ü–µ–Ω—Ç –ø—Ä–æ–ø—É—Å–∫–æ–≤': '{:.2%}'}))
    

# —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –æ–±–∑–æ—Ä–∞ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ –∏ –≤—ã–≤–æ–¥–∞ –ø–µ—Ä–≤—ã—Ö —Å—Ç—Ä–æ–∫

def data_review (data_import):
    
    # –≤—ã–≤–æ–¥–µ–º —Å–ª—É—á–∞–π–Ω—ã–µ 5 —Å—Ç—Ä–æ–∫
    print(display(data_import.sample(5)))
    
    # –≤—ã–≤–µ–¥–µ–º –æ–±—â—É—é –æ–±–∑–æ—Ä–Ω—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ  
    display(pd.DataFrame(data=[data_import.shape[0],
                               data_import.shape[1],
                               data_import.duplicated().sum()],
                         columns=['–í—Å–µ–≥–æ –≤ –¥–∞—Ç–∞—Å–µ—Ç–µ'],
                         index=['–°—Ç—Ä–æ–∫','–°—Ç–æ–ª–±—Ü–æ–≤','–Ø–≤–Ω—ã—Ö –¥—É–±–ª–∏–∫–∞—Ç–æ–≤'])) 
    
    # –≤—ã–≤–æ–¥–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ —Å—Ç–æ–ª–±—Ü–∞—Ö –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞
    data_info(data_import)
```
